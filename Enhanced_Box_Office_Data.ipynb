###### Explanation

Here are key insights based on the correlation matrix:

### 1. **$Domestic and $Foreign Revenue (0.797)**  
There is a strong positive correlation between domestic and foreign revenue, implying that movies performing well in domestic markets also tend to perform well internationally.

### 2. **$Domestic and Vote Count (0.729)**  
A high positive correlation between domestic revenue and vote count suggests that more successful movies garner higher audience engagement, measured through votes.

### 3. **$Foreign and Vote Count (0.665)**  
Foreign revenue also has a positive correlation with vote count, though slightly weaker than domestic revenue. This highlights a general pattern where movies popular in foreign markets attract audience interest.

### 4. **Year and Rating (0.169)**  
A weak positive correlation indicates that newer movies slightly tend to receive better ratings. However, this isn't a strong relationship.

### 5. **Genres and Other Features (Mostly Low Correlations)**  
Genres show negligible correlations with revenue, ratings, and other features, indicating that the genre alone is not a strong predictor for movie success metrics.

### 6. **Original Language and Revenue (Negative Correlations)**  
- **$Domestic (-0.206)** and **$Foreign (-0.140)** revenues negatively correlate with the original language.  
This suggests that movies in English or widely spoken languages tend to perform better in terms of revenue.

### 7. **Production Countries and $Domestic (0.268)**  
There is a moderate positive correlation between production countries and domestic revenue. This could mean that movies produced in popular regions have better access to lucrative domestic markets.

---

### **Key Takeaways**
- Revenue metrics are highly correlated with audience engagement (vote count).  
- Language plays a significant role in predicting revenue, with non-English movies seeing lower revenue on average.  
- Genres have minimal predictive power for financial or audience success.  
- Production countries moderately affect revenue, particularly in the domestic market.

### Model Building

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Read the CSV file containing real estate data into a pandas DataFrame
df = pd.read_csv('/content/data_EnhancedBoxOffice.csv')

# Suppressing warnings
import warnings
warnings.filterwarnings('ignore')  # Ignore warnings to keep the output clean

# Display a random sample of 5 rows from the DataFrame
# This helps in quickly inspecting the structure and content of the data
df.sample(5)
```

#### Regression

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_and_evaluate(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)

    print("Train set prediction:")
    evaluate_model(model, X_train, y_train)

    print("\nTest set prediction:")
    evaluate_model(model, X_test, y_test)

    print("-"*50)

def evaluate_model(model, X, y):
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    mae = mean_absolute_error(y, y_pred)
    r2 = r2_score(y, y_pred)

    print(f"Mean Squared Error: {mse:.4f}")
    print(f"Mean Absolute Error: {mae:.4f}")
    print(f"R-squared: {r2:.4f}")
```

##### 1. Worldwide Revenue Prediction

```python
data = df[['$Worldwide', 'Year', 'Genres', 'Rating', 'Vote_Count', 'Original_Language', 'Production_Countries']]
data.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['$Worldwide'], axis = 1)
y = data['$Worldwide']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.70, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Multi-Layer Perceptron (MLP) Regressor
from sklearn.neural_network import MLPRegressor
mlp_model = MLPRegressor(hidden_layer_sizes=(64, 32), max_iter=1000, random_state=42)
print("MLP Regressor Evaluation:")
train_and_evaluate(mlp_model, X_train, y_train, X_test, y_test)

# Linear Regression
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
print("\nLinear Regression Evaluation:")
train_and_evaluate(lr_model, X_train, y_train, X_test, y_test)

# Support Vector Regressor
from sklearn.svm import SVR
svr_model = SVR(kernel='linear', C=1.0, epsilon=0.2)
print("\nSupport Vector Regressor Evaluation:")
train_and_evaluate(svr_model, X_train, y_train, X_test, y_test)

# K Nearest Neighbour
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors=5)
print("\nK Nearest Neighbour Evaluation:")
train_and_evaluate(knn_model, X_train, y_train, X_test, y_test)

# Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
print("\nRandom Forest Regressor Evaluation:")
train_and_evaluate(rf_model, X_train, y_train, X_test, y_test)

# Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=42)
print("\nDecision Tree Regressor Evaluation:")
train_and_evaluate(dt_model, X_train, y_train, X_test, y_test)

# XG Boost Regressor
from xgboost import XGBRegressor
xgb_model = XGBRegressor(max_depth=8, n_estimators=250, random_state=8030, learning_rate=0.08, n_jobs=5)
print("\nXG Boost Regressor Evaluation:")
train_and_evaluate(xgb_model, X_train, y_train, X_test, y_test)
```

```python
# Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=300,
                                bootstrap=True,
                                max_depth=15,
                                max_features=None,
                                min_samples_leaf=2,
                                min_samples_split=2,
                                random_state=2000)
print("\nRandom Forest Regressor Evaluation:")
train_and_evaluate(rf_model, X_train, y_train, X_test, y_test)
```

```python
# Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(criterion='squared_error',
                              splitter='best',
                              max_depth=None,
                              max_features=None,
                              min_samples_leaf=2,
                              min_samples_split=2,
                              min_impurity_decrease=0.04,
                              random_state=42)
print("\nDecision Tree Regressor Evaluation:")
train_and_evaluate(dt_model, X_train, y_train, X_test, y_test)
```

#### Classification

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

def train_and_evaluate(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)

    print("Train set evaluation:")
    evaluate_model(model, X_train, y_train)

    print("\nTest set evaluation:")
    evaluate_model(model, X_test, y_test)

    print("-" * 50)

def evaluate_model(model, X, y):
    y_pred = model.predict(X)

    acc = accuracy_score(y, y_pred)
    precision = precision_score(y, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y, y_pred, average='weighted')
    f1 = f1_score(y, y_pred, average='weighted')

    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    print("Confusion Matrix:")
    print(confusion_matrix(y, y_pred))
```
